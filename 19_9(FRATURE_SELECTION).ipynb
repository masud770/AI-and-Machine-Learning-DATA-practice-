```python
df= df.dropna()
```

```python
df= df.drop('Unnamed: 0',axis =1)
df
```

```python
from sklearn.preprocessing import LabelEncoder
leb= LabelEncoder()
```

```python
df['city']= leb.fit_transform(df['city'])
```

```python
df['country']= leb.fit_transform(df['country'])
```

```python
df['statezip']= leb.fit_transform(df['statezip'])
```

```python
df
```

```python
x=df.drop('price',axis=1)
x
```

```python
y= df['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
fit_features = SelectKBest(score_func= f_classif, k='all')
fit_features.fit(x,y)
```

```python
fs = pd.DataFrame(fit_features.scores_, index= x.columns,columns=['score values'])
fs
```

```python
fs = fs.nlargest(8,'score values')
fs
```
